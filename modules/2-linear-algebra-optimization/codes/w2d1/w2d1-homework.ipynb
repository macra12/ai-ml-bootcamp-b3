1. Manual Dot Product (No `np.dot`)

```python
import numpy as np
import pandas as pd
```

```python
def manual_dot(a: np.ndarray, b: np.ndarray) -> float:
    """Compute the dot product of two vectors a and b."""
    if a.shape != b.shape:
        raise ValueError("Vectors must be of the same shape.")
    
    result = 0.0
    for i in range(a.shape[0]):
        result += a[i] * b[i]
    
    return result


# Example usage:
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
dot_product = manual_dot(a, b)
print(f"The dot product of {a} and {b} is: {dot_product}")

dot_product_numpy = np.dot(a, b)
print(f"The dot product computed using NumPy is: {dot_product_numpy}")

# Verify that both methods yield the same result
assert np.isclose(dot_product, dot_product_numpy), "The results do not match!"
print("The manual dot product matches the NumPy dot product.")
```

2. Manual Matrix Multiplication (No @ or np.matmul)

```python
def manual_matmul(A: np.ndarray, B: np.ndarray) -> np.ndarray:
    """Compute the matrix multiplication of A and B."""
    if A.shape[1] != B.shape[0]:
        raise ValueError("Number of columns in A must equal number of rows in B.")
    
    result = np.zeros((A.shape[0], B.shape[1]))
    
    for i in range(A.shape[0]):
        for j in range(B.shape[1]):
            for k in range(A.shape[1]):
                result[i, j] += A[i, k] * B[k, j]
    
    return result

# Example usage:
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
matmul_result = manual_matmul(A, B)
print(f"The result of matrix multiplication of A and B is:\n{matmul_result}")

matmul_result_numpy = np.matmul(A, B)
print(f"The result of matrix multiplication computed using NumPy is:\n{matmul_result_numpy}")

# Verify that both methods yield the same result
assert np.allclose(matmul_result, matmul_result_numpy), "The results do not match!"
print("The manual matrix multiplication matches the NumPy matrix multiplication.")
```

3. Batch Linear Prediction

```python
# Format: [Size (sq ft), Number of Bedrooms, Age (years)]
X = np.array([
    [1500, 3, 10], 
    [2500, 4, 5], 
    [1200, 2, 20], 
    [3000, 5, 2],  
    [1800, 3, 15]   
])    
```

```python
# Define a weight vector w and a bias value b
w = np.array([0.20, 15.0, -2.0])
# Bias
b = 50.0

# Compute predictions for all samples using a single matrix multiplication
predictions = X @ w + b
```

```python
print("Real Estate Price Predictions")
for i in range(len(X)):
    size = int(X[i][0])
    beds = int(X[i][1])
    age  = int(X[i][2])
    
    # Multiplying by 1000 to convert the prediction back into full dollar amounts
    predicted_price = predictions[i] * 1000 
    
    print(f"House {i+1} | Size: {size:4d} sqft | Beds: {beds} | Age: {age:2d} yrs | Predicted Price: ${predicted_price:,.2f}")
```

4. Cosine Similarity Application

```python
king = np.array([0.9, 0.3, 0.1])
queen = np.array([0.8, 0.4, 0.2])
apple = np.array([0.1, 0.1, 0.9])
```

```python
def cosine_similarity(vec_a, vec_b):
    dot_product = np.dot(vec_a, vec_b) 
    norm_a = np.linalg.norm(vec_a)    
    norm_b = np.linalg.norm(vec_b)  
    return dot_product / (norm_a * norm_b)
```

```python
sim_king_queen = cosine_similarity(king, queen)
sim_king_apple = cosine_similarity(king, apple)
sim_queen_apple = cosine_similarity(queen, apple)
```

```python
print("Word Vector Semantic Similarity")
print(f"(king, queen) similarity: {sim_king_queen:.4f}")
print(f"(king, apple) similarity: {sim_king_apple:.4f}")
print(f"(queen, apple) similarity: {sim_queen_apple:.4f}")
```
